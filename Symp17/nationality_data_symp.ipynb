```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("2017natData.csv",sep='\t')  #using pandas to convert the csv file into a 2D Data frame
```

```python
df.describe()
```

```python
df = df.replace([" Irish", "Irish ", "irish", "Ireland"], "Irish")
df = df.replace(["UK", "Welsh", " British", "British ", "English", "United Kingdom"], "British")
df = df.replace(["DE", "Deutsch", "German " ," German", "german"], "German")
df = df.replace(["chinese", "CHINESE","China" ], "Chinese")
df = df.replace(["Spain"], "Spanish")
```

```python
summary_inst=df["Nationality"].value_counts() 
sum_df = pd.DataFrame(summary_inst)
sum_df
```

```python
sum_df.to_csv("2017nationality_summary.csv", sep = "\t")
```
