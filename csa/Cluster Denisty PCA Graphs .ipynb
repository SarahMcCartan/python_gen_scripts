```python
f_in=np.loadtxt("310_csa_pca_12py.txt") 
```

```python
x= f_in[:,0]
y= f_in[:,1]
```

```python
xy = np.vstack([x,y])
```

```python
z = gaussian_kde(xy)(xy)
```

```python
# Calculate the point density
xy = np.vstack([x,y])
z = gaussian_kde(xy)(xy)
```

```python
# Sort the points by density, so that the densest points are plotted last
idx = z.argsort()
x, y, z = x[idx], y[idx], z[idx]
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y, c=z, s=30, edgecolor='')
plt.ylim((-2,2) ) #set axes limits
plt.xlim((-2,2.5))
#ax.tick_params(labelsize =50)
ax.set_xlabel("PC 1", fontsize = 20)
ax.set_ylabel("PC 2", fontsize = 20)
#ax.legend("CsA 310K")
#ax.set_title("PCA 310K dCsA, Backbone, )
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets as data
import hdbscan
from sklearn.datasets import make_blobs
%matplotlib inline
```

```python
test_data = ([x, y])
```

```python
plt.scatter(test_data[0], test_data[1])
```

```python
clusterer = hdbscan.HDBSCAN(min_cluster_size=5)
cluster_labels = clusterer.fit_predict(test_data)
```

```python
clusterer.labels_
```

```python
palette = sns.color_palette()
cluster_colors = [sns.desaturate(palette[col], sat)
                  if col >= 0 else (0.5, 0.5, 0.5) for col, sat in
                  zip(clusterer.labels_, clusterer.probabilities_)]
plt.scatter(test_data[0], test_data[1], c=cluster_colors)
```
