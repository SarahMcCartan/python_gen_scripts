```python
#import relevant libraries for plotting
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
from scipy.integrate import simps
from numpy import trapz
import matplotlib.ticker as ticker
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
from matplotlib.backends.backend_pdf import PdfPages
```

```python
import matplotlib.font_manager as fnt
```

```python
%matplotlib inline
```

```python
f_1=np.loadtxt("csa_310_lifetimes_s1.txt") 
f_2=np.loadtxt("csa_310_lifetimes_s2.txt") 
f_3=np.loadtxt("csa_310_lifetimes_s3.txt") 
#f_4=np.loadtxt("csa_450_lifetimes_s4.txt") 
```

```python
s1 = f_1[:,1] # read in  data
s2 = f_2[:,1] 
s3 = f_3[:,1] 
#s4 = f_4[:,1] 
```

```python
sigma = np.std(s1) # standard deviation of distribution
sigma2 = np.std(s2) # standard deviation of distribution
sigma3= np.std(s3) # standard deviation of distribution
#sigma4 = np.std(s4) # standard deviation of distribution

mu = np.mean(s1) # mean of distribution
mu2 = np.mean(s2) # mean of distribution
mu3 = np.mean(s3) # mean of distribution
#mu4 = np.mean(s4) 

print( "std",sigma, sigma2, sigma3)
print("mean", mu, mu2, mu3)
#print( "state4", mu4, sigma4)
```

```python
num_bins = 30  #approx 30 bins found to be optimal for 310 CsA
# the histogram of the data
#normed = True, n is the counts normalized to form a probability density - n/len(x) dbin i.e. integral of hist will be 1

n1, bins, patches = plt.hist(s1, bins=num_bins, normed = True, facecolor='green', alpha=0.5, histtype="bar")
n2, bins2, patches2 = plt.hist(s2, bins = num_bins, normed = True, facecolor='blue', alpha=0.5, histtype = "bar")
n3, bins3,  patches3 = plt.hist(s3, bins = num_bins, normed = True, facecolor = 'red', alpha = 0.5, histtype = "bar")
#n4, bins4,  patches4 = plt.hist(s4, bins = num_bins, normed = True, facecolor = 'orange', alpha = 0.5, histtype = "bar")
```

```python
index = [0]

b1 = np.delete(bins, index)
b2 = np.delete(bins2, index)
b3 = np.delete(bins3, index)
#b4 = np.delete(bins4, index)
```

```python
num_bins = 30

fig1 = plt.figure(1)

# the histogram of the data
#normed = True, n is the counts normalized to form a probability density - n/len(x) dbin i.e. integral of hist willbe 1
plt.plot((0.5*b1), n1, "b", linewidth = 2.0, marker = "o" )

plt.xlabel('T (ns)')
plt.ylabel('P(T)')
plt.legend(["S9"])
#plt.xlim(0, 380)
#plt.ylim(0, 0.085)


fig2 = plt.figure(2)
plt.plot((0.5*b2), n2, "r", linewidth = 2.0, marker = "*" )

plt.xlabel('T (ns)')
#plt.xlim(0, 380)
#plt.ylim(0, 0.085)
plt.legend(["S10"])


fig3 = plt.figure(3)
plt.plot((0.5*b3), n3, "g", linewidth = 2.0, marker = ">" )
plt.xlabel('T (ns)')
plt.ylabel('P(T)')
#plt.xlim(0, 380)
#plt.ylim(0, 0.085)
plt.legend(["S11"])


#fig4 = plt.figure(4)
#plt.plot((0.5*b4), n4, "k", linewidth = 2.0, marker = "s" )


#plt.xlabel('T (ns)')
#plt.ylabel('P(T)')
#plt.xlim(0, 380)
#plt.ylim(0, 0.065)
#plt.legend(["S4"])


plt.show()

#save all figures to 1 pdf
#pp = PdfPages("dCsA_510_Lifetime_Dist.pdf")
#pp.savefig(fig1, dpi = 300, transparent = True)
#pp.savefig(fig2, dpi = 300, transparent = True)
#pp.savefig(fig3, dpi = 300, transparent = True)
#pp.savefig(fig4, dpi = 300, transparent = True)

#pp.close()
```
